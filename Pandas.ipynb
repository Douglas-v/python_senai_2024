```python
#Qual o número total de casos relatados e mortes relacionadas no brasil

dados_covid_copia[['casosAcumulado', 'obitosAcumulado']].tail(1) #filtrando ultima linha do acumulado

total_casos = dados_covid_copia.casosNovos.sum()
total_mortes = dados_covid_copia.obitosNovos.sum()
print('O tatol de mortes foi de',total_mortes,'num total de casos de',total_casos ) #somando colunas
```

```python
#converter object to data
dados_covid_copia['data'] = pd.to_datetime(dados_covid_copia.data)
dados_covid_copia.data
```

```python
#desmembrando as datas convertidas
dados_covid_copia['ano']=pd.DatetimeIndex(dados_covid_copia.data).year
dados_covid_copia['mes']=pd.DatetimeIndex(dados_covid_copia.data).month
dados_covid_copia['dia']=pd.DatetimeIndex(dados_covid_copia.data).day
dados_covid_copia['dia_da_semana']=pd.DatetimeIndex(dados_covid_copia.data).weekday
dados_covid_copia
```

```python
#agrupando dados por mês
dados_covid_copia.groupby('mes')[['casosNovos','obitosNovos']].sum()
```

```python
#agrupando dados por região
dados_covid_copia.groupby('regiao')[['casosNovos','obitosNovos']].sum()
```
