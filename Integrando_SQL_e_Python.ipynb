```python
import sqlite3
```

```python
conn = sqlite3.connect('test.db')
print('Conexão aberta!')
```

```python
conn.execute('''
  CREATE TABLE IF NOT EXISTS Empregado(
    matricula integer PRIMARY KEY AUTOINCREMENT,
    nome string,
    CPF integer
  )
''')
conn.commit()
print('Tabela criada com sucesso.')
```

```python
conn.execute("INSERT INTO Empregado VALUES(1,'thiago medeiros',000)")
conn.execute("INSERT INTO Empregado VALUES(2,'miguel medeiros',123)")
conn.execute("INSERT INTO Empregado VALUES(3,'thalita barros',454)")
conn.execute("INSERT INTO Empregado VALUES(4,'maria de fatima',644)")

conn.commit()
```

```python
cursor = conn.execute('''
select nome, cpf
from empregado
''')

for row in cursor:
  print(row)
```

```python
import pandas as pd

qry = '''
select *from Empregado
'''
df = pd.read_sql_query(qry,conn)
df.head()
```

```python
conn.execute('''

create table if not exists df(
  custID integer primary key autoincrement,
  name string
)'''
)
conn.commit()
print('tabela criada com sucesso.')
```
