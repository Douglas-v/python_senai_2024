```python
#importando as bibliotecas
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#configurando o estilo dos gráficos
sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 14
matplotlib.rcParams['figure.figsize']=(9,5)
matplotlib.rcParams['figure.facecolor']='#00000000'
```

```python
#dados
ano = range(2011,2019)
messi = [59,91,45,58,52,59,54,51]
cr7 = [60,63,69,91,57,55,53,49]

#primeiro gráfico de linha
plt.plot(ano,messi,'b-x',linewidth=2,markersize=6,markeredgewidth=2,markeredgecolor='navy')

#segundo grafico de linha
plt.plot(ano,cr7,'r--o',linewidth=2,markersize=6,)

#titulo
plt.title('Messi VS CR7')

#legenda
plt.legend(['Messi','cr7'])

#eixos
plt.xlabel('Ano')
plt.ylabel('Gols')

#salvar imagem
plt.savefig('grafico.png')
```

```python
import pandas as pd
iris = pd.read_csv('/content/iris.csv')
iris
```

```python
#scatter plot
sns.scatterplot(data=iris, x='sepal_length', y="sepal_width", hue='species')
plt.title('Flores')
```

```python
#gráfico de distribuição
plt.title('Distribuição Sepal width')
sns.distplot(iris.sepal_width, kde=False)
```

```python
#salvando imagem
swarm_plot = sns.distplot(iris.sepal_width, kde=False)
fig = swarm_plot.get_figure()
fig.savefig('out.png')
```

```python
#importando datasets nativos da biblioteca Seaborn
sns.get_dataset_names()
```

```python
#Carregar dados de voo e fazer um "pivot" para valores especificos
voos = sns.load_dataset('flights').pivot_table(index='month',columns='year',values='passengers')

#titulo
plt.title('Numero de passageiros (1000)')

#desnehar o grafico
sns.heatmap(voos,
            fmt='d',
            annot=True,
            linewidths=.5,
            cmap='Reds',
            annot_kws={'fontsize':13})
```

```python
#contour plot
plt.title('Flores')
sns.kdeplot(data=iris,x='sepal_length',y='sepal_width',shade=True,shade_lowest=False)
```

```python
#contour plot
plt.title('Flores')
sns.kdeplot(data=iris,x='sepal_length',y='sepal_width',hue='species',fill=True)
```
