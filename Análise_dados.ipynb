```python
def split_multicolumn(col_series):
    result_df = col_series.to_frame()
    options = []
    # percorrer as colunas
    for idx, value in result_df[col_series.notnull()].iterrows():
        # Divida cada valor em uma lista de opções
        for option in value[0].split(';'):
            # Adicione a opção como uma coluna ao resultado
            if option not in result_df.columns:
                options.append(option)
                result_df[option] = False
            # Marque o valor na coluna de opções como True
            result_df.at[idx, option] = True
    return result_df[options]
```

```python
linguagem_trabalham = split_multicolumn(dados.LanguageHaveWorkedWith)
```

```python
linguagem_trabalham
```

```python
#verificando a media de todos os valores e organizar de forma descrecente
linguagem_trabalham = linguagem_trabalham.mean().sort_values(ascending=False) * 100
```

```python
linguagem_trabalham
```

```python
#gerar um gráfico
plt.figure(figsize=(12,12))
sns.barplot(x=linguagem_trabalham, y=linguagem_trabalham.index)
plt.title('Qual linguagem você usou esse ano')
plt.xlabel('Quantidade')
```

```python
linguagem_intere = split_multicolumn(dados.LanguageWantToWorkWith)
linguagem_intere_percent = linguagem_intere.mean().sort_values(ascending=False) * 100
linguagem_intere_percent
```

```python
#gerar um gráfico
plt.figure(figsize=(12,12))
sns.barplot(x=linguagem_intere_percent, y=linguagem_intere_percent.index)
plt.title('Qual linguagem as pessoas querem estudar no próximo ano')
plt.xlabel('Quantidade')
```
